```python
import numpy as np
import matplotlib.pyplot as plt
import math

#Generate an array of values from 0 to 2.5 (not including 2.5) with step size 0.1
t = np.arange(0, 2.5, 0.1)

# Calculate y1 as the sine of (π * t) for each value in t
y1 = np.sin(math.pi * t)
# Calculate y2 as the sine of (π * t + π/2)
y2 = np.sin(math.pi * t + math.pi / 2)

# Create a figure
figure = plt.figure()

# Add a subplot
axes = figure.add_subplot()

# Plot t vs y1 as blue star markers
axes.plot(t,y1,'b*')
# Plot t vs y2 with green triangle
axes.plot(t,y2,'g^')
plt.show()
```

```python
# Generate an array of values from 0 to 1 (not including 1) with step size 0.05
t = np.arange(0, 1, 0.05)

# Calculate y1 as the sine of (t * 2π) for each value in t
y1 = np.sin(t * np.pi * 2)
# Calculate y2 as the cosine of (t * 2π) for each value in t
y2 = np.cos(t * np.pi * 2)

# Create a figure
figure = plt.figure()

# Add the first subplot (1 row, 2 columns, position 1) for the sine wave
sin1 = figure.add_subplot(1,2,1)
# Add the second subplot (1 row, 2 columns, position 2) for the cosine wave
cos1 = figure.add_subplot(1,2,2)

# Plot t vs y1 with blue dash-dot lines
sin1.plot(t, y1, 'b-.')
# Plot t vs y2 with red dashed lines
cos1.plot(t, y2, 'r--')

plt.show()
```
