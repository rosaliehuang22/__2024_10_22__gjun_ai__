```python
import random
import matplotlib.pyplot as plt

fig = plt.figure()

days = [num for num in range(1,31)]
celsius = [round(random.uniform(28.5, 35.2),1) for _ in range(1,31)]

fig = plt.figure(figsize=(10, 6))
plt.plot(days, celsius, 'o--r')
plt.title('july weather')
plt.xlabel('Days')
plt.ylabel('Celcius')
plt.xticks(days)
plt.yticks(range(28, 36, 1))
plt.grid('y')  # Gridlines only for y-axis

plt.show()



```
