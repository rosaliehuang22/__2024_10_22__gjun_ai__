```python
import random

# Define BMI class
class BMI():
    def __init__(self,name,height,weight):
        #Initialize name, height and weight information
        self.name = name
        self.height = height
        self.weight = weight

    def get_bmi(self):
        #Calculate BMI
        return round(self.weight / (self.height / 100) ** 2,2)
    
    def get_status(self):
        #Determine body status based on BMI
        bmi = self.get_bmi()
        if bmi >=35:
            status = "您的體重:重度肥胖"
        elif bmi >= 30:
            status = "您的體重:中度肥胖"
        elif bmi >= 27:
            status = "您的體重:輕度肥胖"
        elif bmi >= 24:
            status = "您的體重:過重"
        elif bmi >= 18.5:
            status = "您的體重:正常範圍"
        else:
            status = "您的體重:體重過輕"
        return status
    
    def get_info(self) -> dict:
        return {
            '姓名': self.name,
            '身高': self.height,
            '體重': self.weight,
            'BMI': self.get_bmi(),
            '狀態': self.get_status()
        }

# Function to get list of names from file
def get_name(nums:int=2)->list[str]:
    with open('names.txt',mode='r',encoding='utf-8') as file:
        names_str = file.read()
    names = names_str.split()
    names = random.choices(names,k=nums)
    return  names

# Hàm tạo danh sách đối tượng BMI
def generate_bmi(names: list[str]) -> list[BMI]:
    students: list[BMI] = []
    for name in names:
        height = random.randint(140, 190)
        weight = random.randint(50, 110)
        bmi = BMI(name=name, height=height, weight=weight)
        students.append(bmi)
    return students

if __name__ == "__main__":
    nums = int(input('請輸入人數: '))
    names: list[str] = get_name(nums=nums)
    students: list[BMI] = generate_bmi(names=names)

    for student in students:
        student = student.get_info()
        for key, value in student.items():
            print(f'{key}: {value}')
        print("==================")
```
