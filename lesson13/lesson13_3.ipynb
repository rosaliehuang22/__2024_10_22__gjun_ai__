```python
import numpy as np
import matplotlib.pyplot as plt

# Define categories (x-axis labels) and their corresponding values (y-axis values)
index = ['apple', 'banana', 'orange', 'tomoto', 'guava']
values = [5,7,3,4,6]

figure = plt.figure()
axes = figure.add_subplot()
axes.bar(index,values)

# Print the positions of the x-ticks on the plot (locations of each bar)
print(axes.get_xticks())
# Print the labels of the x-ticks (text labels for each bar)
print(axes.get_xticklabels())

plt.show()
```

```python
# Tạo một danh sách các vị trí trên trục x để đặt các cột
index = [3, 4, 5, 6, 7]
# Tạo một danh sách tên chủ đề tương ứng với từng cột
subject = ['apple', 'banana', 'orange', 'tomato', 'guava']
# Tạo một danh sách chiều cao của các cột
values = [5000, 6000, 7000, 8000, 9000]

figure = plt.figure()
axes = figure.add_subplot()
axes.bar(index, values)

# Thiết lập các vị trí trên trục x và gán nhãn cho từng vị trí bằng các giá trị trong `subject`
axes.set_xticks(index,labels=subject)

# Lấy các giới hạn hiện tại của trục x và y
xmin, xmax, ymin, ymax = axes.axis()

# Điều chỉnh giới hạn trục x để có khoảng trống thêm bên trái và bên phải
xmin -= 0.2 # Giảm giá trị tối thiểu của trục x
xmax += 0.2 # Tăng giá trị tối đa của trục x

# Điều chỉnh giới hạn trục y để có khoảng trống thêm phía dưới và phía trên
ymin = 3000  # Đặt giá trị tối thiểu của trục y là 3000
ymax += 1000  # Tăng giá trị tối đa của trục y thêm 1000

# Đặt lại các giới hạn cho trục x và y với các giá trị đã điều chỉnh
axes.axis(xmin = xmin, xmax = xmax, ymin = ymin, ymax = ymax)

# Thêm chú thích vào mỗi cột, hiển thị giá trị của cột đó
for idy, x in enumerate(index):
    axes.annotate(str(values[idy]),(x,values[idy])) # Thêm văn bản chú thích là giá trị của cột tại (x, values[idy])

plt.show()
```
