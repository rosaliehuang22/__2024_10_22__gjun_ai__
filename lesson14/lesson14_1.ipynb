```python
#統計使用實體方法
#tạo mảng 2D có size 10 hàng + 5 cột
arr = np.random.randint(50,101,size=(10,5))
#tính tổng tất cả các phần tử trong mảng
np.sum(arr)
#(axis=1) = tổng 5 giá trị của cột
np.sum(arr,axis=0)
#(axis=1) = tổng 10 giá trị của hàng
np.sum(arr,axis=1)
```

Cách tính Cumsum

Vị trí 0: 0 (chỉ có 0)\
Vị trí 1: 0 + 1 = 1\
Vị trí 2: 0 + 1 + 2 = 3\
...
Vị trí 7: 0 + 1 + 2 + 3 + 4 + 5 + 6 + 7 = 28

```python
#cumsum = tính tổng tích lũy của các phần tử trong mảng
arr = np.array([0,1,2,3,4,5,6,7])
arr.cumsum()
```

```python
#tạo mảng 1D với 30 số nguyên ngẫu nhiên
oneStore = np.random.randint(100,10000,size=(30,))
oneStore
```

```python
import matplotlib.pyplot as plt

#Figure với canvas = 10 đơn vị rộng, 5 đơn vị cao
fig = plt.figure(figsize=(10,5))

#Thêm trục
#.add_subplot(nrows, ncols, index)
axes1 = fig.add_subplot(2,1,1)

#Vẽ biểu đồ đường trên trục axes1
axes1.plot(range(1,31), oneStore,'b-')

#Đặt giá trị đánh dấu trên trục x tại các vị trí từ 1 đến 30
axes1.set_xticks(range(1,31))

axes2 = fig.add_subplot(2,1,2)
axes2.plot(range(1,31),oneStore.cumsum(),'r-')
axes2.set_xticks(range(1,31))
plt.show()
```

```python
five_store = np.random.randint(100,10000, size = (5,30))
five_store_cumsum = five_store.cumsum(axis=1)
five_store_cumsum
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(10,5))
axes = fig.add_subplot()
axes.set_xticks(range(1,31))
#Ve grid theo x
axes.grid(axis='x')
for one in five_store_cumsum:
    axes.plot(range(1,31),one)

plt.show()
```
